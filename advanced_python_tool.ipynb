```python
import numpy as np

# Part 1 – 1-D array (1 to 20)
print("Part 1 – 1-D Array")
arr1 = np.arange(1, 21)
print("Array:", arr1)

print("Sum:", arr1.sum())
print("Mean:", arr1.mean())
print("Median:", np.median(arr1))
print("Standard Deviation:", arr1.std())

indices_gt_10 = np.where(arr1 > 10)[0]
print("Indices of elements > 10:", indices_gt_10)
print()

# Part 2 – 2-D array (4×4) with values 1 to 16
print("Part 2 – 2-D Array (4×4)")
arr2 = np.arange(1, 17).reshape(4, 4)
print("2-D Array:\n", arr2)

arr2_T = arr2.T
print("Transpose of the array:\n", arr2_T)

print("Row-wise sum:", arr2.sum(axis=1))
print("Column-wise sum:", arr2.sum(axis=0))
print()

# Part 3 – two random 3×3 arrays
print("Part 3 – Random 3×3 Arrays")
a = np.random.randint(1, 21, size=(3, 3))
b = np.random.randint(1, 21, size=(3, 3))
print("Array A:\n", a)
print("Array B:\n", b)

print("Addition:\n", a + b)
print("Subtraction:\n", a - b)
print("Multiplication:\n", a * b)
print("Dot product (A @ B):\n", a @ b)
print()

# Part 4 – reshape and slice
print("Part 4 – Reshape and Slice")
arr4 = np.arange(1, 13).reshape(3, 4)
print("Original 3×4 Array:\n", arr4)

sliced = arr4[:2, -2:]
print("Sliced (first 2 rows, last 2 columns):\n", sliced)
```

```python
import pandas as pd


data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Age': [24, 27, 22, 32, 29],
    'Department': ['HR', 'Finance', 'IT', 'Marketing', 'HR'],
    'Salary': [45000, 54000, 50000, 62000, 47000]
}
df = pd.DataFrame(data)

# Part 1a – first five rows
print("Part 1a – First five rows")
print(df)

# Part 1b – summary stats for Age & Salary
print("\nPart 1b – Summary statistics (Age & Salary)")
print(df[['Age', 'Salary']].describe())

# Part 1c – average Salary of HR employees
print("\nPart 1c – Average Salary (HR)")
hr_avg_salary = df[df['Department'] == 'HR']['Salary'].mean()
print("Average Salary (HR):", hr_avg_salary)

# Part 2 – add Bonus column (10 % of Salary)
print("\nPart 2 – Bonus column")
df['Bonus'] = df['Salary'] * 0.10
print(df)

# Part 3 – employees aged 25–30
print("\nPart 3 – Employees aged 25 – 30")
filtered_df = df[(df['Age'] >= 25) & (df['Age'] <= 30)]
print(filtered_df)

# Part 4 – average Salary by Department
print("\nPart 4 – Average Salary by Department")
grouped_salary = df.groupby('Department')['Salary'].mean()
print(grouped_salary)

# Part 5 – sort by Salary (ascending) and save to CSV
print("\nPart 5 – Sorted by Salary (Ascending)")
sorted_df = df.sort_values(by='Salary')
print(sorted_df)

sorted_df.to_csv("sorted_employment.csv", index=False)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# line plot
x_vals = [1, 2, 3, 4, 5]
y_vals = [10, 15, 25, 30, 50]

plt.figure(figsize=(8, 5))
plt.plot(x_vals, y_vals, marker="o", linewidth=2)
plt.title("Growth Trend")
plt.xlabel("Index")
plt.ylabel("Value")
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# Bar chart
students = ["John", "Jane", "Alice", "Bob"]
marks = [75, 85, 60, 90]
colors = plt.cm.Set2(range(len(students)))  # quick pastel palette

plt.figure(figsize=(8, 5))
plt.bar(students, marks, color=colors)
plt.title("Marks Scored by Students")
plt.ylabel("Score")
plt.ylim(0, 100)
plt.tight_layout()
plt.show()

# Pie chart 
regions = ["North America", "Europe", "Asia", "Others"]
revenue = [45, 25, 20, 10]
explode = [0.1 if r == max(revenue) else 0 for r in revenue]  # highlight top slice

plt.figure(figsize=(6, 6))
plt.pie(
    revenue,
    labels=regions,
    explode=explode,
    autopct="%1.1f%%",
    startangle=160,
    shadow=True,
)
plt.title("Company Revenue Distribution by Region")
plt.tight_layout()
plt.show()

# Histogram 
np.random.seed(42)
random_data = np.random.randint(1, 101, 1000)

plt.figure(figsize=(8, 5))
plt.hist(random_data, bins=20, edgecolor="black", alpha=0.8)
plt.title("Distribution of Random Integers (1–100)")
plt.xlabel("Integer Value")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```
