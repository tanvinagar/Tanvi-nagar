```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load and clean the data
data = pd.read_csv("austin_weather.final.csv")

# Drop unnecessary columns if they exist
cols_to_drop = ["Events", "Date", "SeaLevelPressureLowInche"]
data = data.drop(columns=[c for c in cols_to_drop if c in data.columns], errors='ignore')

# Replace 'T' and '-' with 0.0
data = data.replace('T', 0.0)
data = data.replace('-', 0.0)

# Convert everything to numeric (in case any column has object dtype after replacements)
data = data.apply(pd.to_numeric, errors='coerce')

# Drop rows with missing values
data = data.dropna()

# Define features and target
X = data.drop(['PrecipitationSumInches'], axis=1)
Y = data["PrecipitationSumInches"].values.reshape(-1, 1)

# Fit the regression model
clf = LinearRegression()
clf.fit(X, Y)

# Define days and day_index for plotting
days = list(range(Y.size))
day_index = [798]  # You can change this as needed

# Plot precipitation trend
print("The precipitation trend graph:")
plt.figure(figsize=(8, 4))
plt.scatter(days, Y, color='g')
plt.scatter(day_index, Y[day_index], color='r')
plt.title("Precipitation level")
plt.xlabel("Days")
plt.ylabel("Precipitation in inches")
plt.show()

# Select attributes to visualize
x_vis = X.filter([
    'TempAvgF', 'DewPointAvgF', 'HumidityAvgPercent',
    'SeaLevelPressureAvgInches', 'VisibilityAvgMiles', 'WindAvgMPH'
])

# Plot 6 attributes in 3x2 grid
print("Precipitation vs. attribute trends:")

fig = plt.figure(figsize=(12, 10))

for i, col in enumerate(x_vis.columns):
    plt.subplot(3, 2, i + 1)
    plt.scatter(days, x_vis[col], color='g', alpha=0.6)
    plt.scatter(day_index, x_vis[col].iloc[day_index], color='r')
    plt.title(col)

plt.tight_layout()
plt.show()
```
